```python
import scipy.io.wavfile
from CC_utils import CQ_Spectrogram, GC_Spectrogram, CQCC, GFCC
```

```python
fs, sig = scipy.io.wavfile.read('samples/blues.00009.wav')
sp = CQ_Spectrogram()
spectogram = sp.cqt_transform(sig)
cq = CQCC()
cqcc_our_blues = cq.cqcc(spectogram)

gf_sp = GC_Spectrogram()
spectogram2, ft = gf_sp.spectrogram(sig, fs)
gf = GFCC()
gfcc_our_blues = gf.gfcc(spectogram2)

fs, sig = scipy.io.wavfile.read('samples/classical.00095.wav')
sp = CQ_Spectrogram()
spectogram = sp.cqt_transform(sig)
cq = CQCC()
cqcc_our_classical = cq.cqcc(spectogram)

gf_sp = GC_Spectrogram()
spectogram2, ft = gf_sp.spectrogram(sig, fs)
gf = GFCC()
gfcc_our_classical = gf.gfcc(spectogram2)

fs, sig = scipy.io.wavfile.read('samples/disco.00003.wav')
sp = CQ_Spectrogram()
spectogram = sp.cqt_transform(sig)
cq = CQCC()
cqcc_our_disco = cq.cqcc(spectogram)

gf_sp = GC_Spectrogram()
spectogram2, ft = gf_sp.spectrogram(sig, fs)
gf = GFCC()
gfcc_our_disco = gf.gfcc(spectogram2)

```

```python

from spafe.features.cqcc import cqcc
from spafe.features.gfcc import gfcc
fs, sig = scipy.io.wavfile.read('samples/blues.00009.wav')

cqcc_spafe_blues = cqcc(sig=sig, fs=fs, pre_emph=True)
gfcc_spafe_blues = gfcc(sig=sig, fs=fs, pre_emph=True)

fs, sig = scipy.io.wavfile.read('samples/classical.00095.wav')

cqcc_spafe_classical = cqcc(sig=sig, fs=fs, pre_emph=True)
gfcc_spafe_classical = gfcc(sig=sig, fs=fs, pre_emph=True)

fs, sig = scipy.io.wavfile.read('samples/disco.00003.wav')

cqcc_spafe_disco = cqcc(sig=sig, fs=fs, pre_emph=True)
gfcc_spafe_disco = gfcc(sig=sig, fs=fs, pre_emph=True)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(cqcc_our_blues.T, aspect='auto', origin='lower')
axes.flat[1].imshow(cqcc_spafe_blues.T, aspect='auto', origin='lower')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(cqcc_our_classical.T, aspect='auto', origin='lower')
axes.flat[1].imshow(cqcc_spafe_classical.T, aspect='auto', origin='lower')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(cqcc_our_disco.T, aspect='auto', origin='lower')
axes.flat[1].imshow(cqcc_spafe_disco.T, aspect='auto', origin='lower')
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(gfcc_our_blues.T, aspect='auto', origin='lower')
axes.flat[1].imshow(gfcc_spafe_blues.T, aspect='auto', origin='lower')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(gfcc_our_classical.T, aspect='auto', origin='lower')
axes.flat[1].imshow(gfcc_spafe_classical.T, aspect='auto', origin='lower')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes.flat[0].imshow(gfcc_our_disco.T, aspect='auto', origin='lower')
axes.flat[1].imshow(gfcc_spafe_disco.T, aspect='auto', origin='lower')
```
